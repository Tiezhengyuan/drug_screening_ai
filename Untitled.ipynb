```python
  1. System Architecture

  [Data Layer]
  ├── MFI CSVs (Treatment, Concentration, Control)
  ├── Pooling CSVs (Well assignments, Library IDs)
  └── README (Experimental protocol, metadata)

  [Knowledge Layer]
  ├── PRISM PDF Publications (Background, Methodology, Prior Screens)
  ├── PDF Extractor (PyPDF2/pdfplumber)
  ├── Embedding Model (sentence-transformers, local)
  └── Vector DB (ChromaDB, local)

  [AI Agent Core]
  ├── Orchestrator: LangChain
  ├── LLM: Qwen (via Ollama, 100% local)
  ├── Tools:
  │   ├── `query_data()` → Returns processed PRISM metrics
  │   ├── `query_knowledge()` → Retrieves relevant PRISM paper context
  │   └── `analyze_response()` → Statistical & pattern analysis
  └── Memory: LangChain ConversationBufferMemory
```

```python
2. Step-by-Step Implementation Plan
  
  Phase 1: Environment & Local Model Setup

  1. Install core dependencies: langchain, langchain-community, ollama, pandas, pdfplumber, chromadb, sentence-transformers.
  2. Pull and run Qwen locally: ollama run qwen2.5:14b (or larger depending on VRAM).
  3. Verify local LLM connectivity via Python.

  Phase 2: Data Ingestion Pipeline

  1. Load MFI CSVs into Pandas DataFrames.
  2. Parse experimental metadata from README (normalize plate layouts, drug names, concentrations).
  3. Cross-reference pooling CSVs to map compounds to well positions.
  4. Clean data: normalize MFI values, handle missing wells, calculate fold-changes vs controls.

  Phase 3: Knowledge Base Construction

  1. Extract text from PRISM PDFs.
  2. Chunk text (e.g., 500 tokens, 50 overlap).
  3. Generate embeddings using sentence-transformers (e.g., all-MiniLM-L6-v2).
  4. Store in ChromaDB with metadata (title, year, section).

  Phase 4: AI Agent Development

  1. Define LangChain tools for data querying & knowledge retrieval.
  2. Build a ReAct agent that decides whether to:
    - Query the PRISM dataset for metrics.
    - Retrieve background from PDFs.
    - Run analysis or synthesize findings.
  3. Wrap in a conversational interface.

  Phase 5: Integration & Output

  1. Chain ingestion → knowledge → agent.
  2. Provide structured outputs: drug efficacy rankings, mechanistic hypotheses, data quality flags.
  3. Export results to CSV/JSON + markdown report.

```
